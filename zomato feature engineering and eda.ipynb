```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings                        
```

```python
sns.set()
plt.figure(figsize=(12,4))
a=sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings,palette=['blue','red','orange','yellow','cyan','green'])
for ax in a.containers:
    a.bar_label(ax)
```

```python
#countplot
sns.set()
a=sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','cyan','green'])
for ax in a.containers:
    a.bar_label(ax)
```
